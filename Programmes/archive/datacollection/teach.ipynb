```python
class ppl:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
    
    def propose(self, name):
        print(f"{self.name} proposes to {name}")
```

```python
mianbao = ppl("mianbao", 27)
PhD = ppl("PhD", 30)
mianbao.propose(PhD.name)
```

```python
class Dog:
    species = "Dog"

    def __init__(self, name):
        self.name = name

d1 = Dog("Buddy")
d2 = Dog("Lucy")

d1.name = "Max"

print(d1.name)      # Max
print(d2.name)      # Lucy
print(d1.species)   # Animal
print(d2.species)   # Animal
```
